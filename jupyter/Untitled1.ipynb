```python
from pprint import pprint
```

```python
import requests
import json
site = 'https://bank.gov.ua/NBUStatService/v1/statdirectory/exchange?date=20210609&json'
# site = "https://api.privatbank.ua/p24api/pubinfo?exchange&coursid=5"
url = requests.get(site)
htmltext = url.text
text = json.loads(htmltext)
pprint(text)
# print(htmltext)
```

```python
for i in text:
#     print(i['txt'])
    if i['txt'] == 'Долар США':
        print(i['rate'])
```
